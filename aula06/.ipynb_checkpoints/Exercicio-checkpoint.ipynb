```python
class Stack:
    def __init__(self):
        self.itens = []
    
    def is_empty(self):
        return self.itens == []
    
    def push(self, item):
        self.itens.append(item)
        print(f"PUSH {item}")

    def pop(self):
        print("POP")
        return self.itens.pop()
    
    def peek(self):
        return self.itens[-1]
    
    def size(self):
        return len(self.itens)
    
    def print_stack(self):
        print(self.itens)
```

```python
resultado = Stack()
```

```python
number = 45
resto = 0

while number > 0:
    resto = number % 2
    resultado.push(resto)
    number // 2
    
```
