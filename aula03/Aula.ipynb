```python
def find_element(list, key):
    try:
        for i in range (len(list)):
            if list[i] == key:
                print(f"valor encontrado! {list[i]} na posição {i}")
    except:
        return -1
```

```python
list1 = [11,  18, 3, 5, 8]
```

```python
find_element(list1, 8)
```

```python
def find_element2(list, key):
    init = 0 
    try:
        while init < len(list):
            if list[init] == key:
                return init
            init+=1
    except:
        return -1
```

```python
x = find_element2(list1, 8)
```

```python
if x == -1:
    print(f"Valor não encontrado na lista")
else:
    print(f"Valor encontrado na posição {x}")
```

```python
def buscar_cliente(nome, lista_clientes):
    for i in range (len(lista_clientes)):
        if lista_clientes[i] == nome:
            return i
    return -1
```

```python
lista_clientes = ["Luiz", "André", "Emanuel", "Frederico"]
busca1 = "Emanuel"
busca2 = "Fred"
```

```python
pos_cliente1 = buscar_cliente(busca1, lista_clientes)
pos_cliente2 = buscar_cliente(busca2, lista_clientes)

if pos_cliente1 != -1:
    print(f"Cliente {busca1} encontrado na posição {pos_cliente1}")
else:
    print("Cliente não encontrado")

if pos_cliente2 != -1:
    print(f"Cliente encontrado na posição {pos_cliente2}")
else:
    print(f"Cliente {busca2} não encontrado")
```
