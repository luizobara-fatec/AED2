### Projeto 1: A Agenda de Contatos:
• **Contexto:** Você precisa criar um sistema simples para armazenar e gerenciar contatos (nome,
telefone e e-mail).

• **Problema:** Como você pode adicionar novos contatos, buscar um contato pelo nome e listar
todos os contatos?

• **Indicação de como resolver:** Pensar em uma estrutura que permita associar informações
(telefone e e-mail) a um nome de forma eficiente.

```python
agenda = {}
```

```python
def add_contact(name, phoneNumber, email):
    agenda[name] = {
        "telefone": phoneNumber,
        "email": email
    }
```

```python
add_contact("Luiz", "1234-5678", "luiz@gmail.com")
```

```python
add_contact("Julia", "9876-5432", "julia@gmail.com")
```

```python
agenda
```

```python
def find_contact(name):
    contact = agenda.get(name)
    
    if contact:
        print(f"Nome: {name}")
        print(f"Telefone: {contact['telefone']}")
        print(f"Email: {contact['email']}")
    else:
        print("Contato não encontrado.")
```

```python
find_contact("Luiz")
```

```python
def all_contacts():
    if not agenda:
        print("Agenda vazia.")
        return
    
    for name, data in agenda.items():
        print(f"\nNome: {name}")
        print(f"Telefone: {data['telefone']}")
        print(f"Email: {data['email']}")
```

```python
all_contacts()
```
