```python
class Stack:
    def __init__(self):
        self.itens = []
    
    def is_empty(self):
        return self.itens == []
    
    def push(self, item):
        self.itens.append(item)

    def pop(self):
        return self.itens.pop()
    
    def peek(self):
        return self.itens[-1]
    
    def size(self):
        return len(self.itens)
    
    def print_stack(self):
        print(self.itens)
```

```python
def dec_binario(numero):
    resultado = Stack()
    
    while numero > 0:
        resto = numero % 2
        resultado.push(resto)
        numero = numero // 2

    binario = ''
    while not resultado.is_empty():
        binario = binario + str(resultado.pop())
    return binario
```

```python
dec_binario(45)
```
